```python
# Download the iris dataset from a public URL
!wget https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data -O iris.csv

# Import necessary libraries
import pandas as pd
from sklearn.linear_model import LogisticRegression

#Loading the Dataset and Displaying the first few rows
iris_data = pd.read_csv('iris.csv', names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
print(iris_data.to_string())
```

```python
#Split the data into features (X) and labels (Y)
x = iris_data.drop(columns=['class'])
y = iris_data['class']

print (x.head())
print (y.head())
```

```python
#create a ml model
model = LogisticRegression()
```

```python
#train the model
model.fit(x.values, y)
```

```python
#predict using the trained model
predictions = model.predict([[7.2, 3.2, 6.0, 1.8]])
```

```python
#print the predictions
print(predictions)
```
